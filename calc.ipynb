```python
import numpy as np
import re
import matplotlib.pyplot as plt
```

```python
path = "../CLionProjects/Sirius/f10_512.txt"

pair = re.compile(b'\(([^,\)]+),([^,\)]+)\)')
def parse_pair(s):
    return complex(*map(float, pair.match(s).groups()))
```

```python
f = np.loadtxt(path, dtype=complex,  converters={0:parse_pair})
f.shape
```

```python
pair = re.compile('\(([^,\)]+),([^,\)]+)\)')
z = []
with open("../CLionProjects/Sirius/matrix10_512.txt", "r") as f:
    i = 0
    for line in f:
        b = line.split(" ")
        i += 1
        z += b[:-1]
```

```python
z = [parse_pair(s) for s in z]
```

```python
z = np.array(z, dtype=complex).reshape(768, 768)
```

```python
z.shape, f.shape
```

```python
x = np.linalg.solve(z, f)
x
```

```python
np.savetxt("../CLionProjects/Sirius/results/k=10/j10_768.txt", x)
```
